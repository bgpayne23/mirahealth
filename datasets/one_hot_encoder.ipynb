```python
#Loading Libraries
import pandas as pd
import numpy as np
import requests
import io
from sklearn.preprocessing import OneHotEncoder
```

```python
# Downloading the csv file from your GitHub account
url = "https://raw.githubusercontent.com/bgpayne23/mirahealth/main/datasets/mira_data_v2.csv" # Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content
# Reading the downloaded content and turning it into a pandas dataframe
df = pd.read_csv(io.StringIO(download.decode('utf-8')))
# Printing out the first 5 rows of the dataframe
df.drop(['Unnamed: 0'], axis=1, inplace=True)
datatypes = df.dtypes
datatypes
```

```python
#place objects in list
objects = list(df.select_dtypes(['object']).columns)
objects
```

```python
one_hot_cols = pd.get_dummies(df[objects])
one_hot_cols
```

```python
#Redo the dataset
df.drop(objects, axis=1, inplace=True)
result = pd.concat([df, one_hot_cols], axis=1, join='inner')
display(result)
```

```python
#Save as the new training dataset
result.to_csv('mira_data_one_hot.csv')
```
